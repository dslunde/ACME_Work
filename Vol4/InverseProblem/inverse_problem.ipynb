```python
import numpy as np
from scipy.optimize import minimize
from matplotlib import pyplot as plt
```

### Problem 1

```python
u = lambda x : x**2 + x/2 + 5/16
u_prime = lambda x : 2*x + 1/2
f = lambda x : -1 if x > 1/2 else -6*x**2 + 3*x - 1
f_int = lambda x : -2*x**3 + 3/2*x**2 - x if x < 1/2 else 1/8 - x
c0 = 3/8
c1 = 5/4
xj = np.arange(0,1.1,.1)
def F(alpha) :
    return sum([((c0-f_int(xj[i]))/alpha[i] - u_prime(xj[i]))**2 for i in range(len(xj))])
guess = np.ones_like(xj)
alphas = minimize(F,guess)
```

```python
plt.plot(xj,alphas.x,'.-')
plt.title('a(x)')
plt.axis([0,1,.45,.75])
plt.show()
```
