## Problem 1

```python
def plot_complex(f, xbounds, ybounds, res=401):

    """Plot the complex function f.
    INPUTS:
        f - A function handle. Should represent a function from C to C.
        xbounds - A tuple (xmin, xmax) describing the bounds on the real part of the domain.
        ybounds - A tuple (ymin, ymax) describing the bounds on the imaginary part of the domain.
        res - A scalar that determines the resolution of the plot (number of points
            per side). Defaults to 401.
    
    OUTPUTS:
        graph of the function f(z)
    """
    #raise NotImplementedError("Problem 1 Incomplete")
    x = np.linspace(xbounds[0],xbounds[1],res)
    y = np.linspace(ybounds[0],ybounds[1],res)
    X,Y = np.meshgrid(x,y)
    Z = f(X + 1j*Y)
    plt.pcolormesh(X,Y,np.angle(Z),cmap='hsv',vmin=-np.pi,vmax=np.pi)
    plt.axis([xbounds[0],xbounds[1],ybounds[0],ybounds[1]])
    plt.show()
```

Plot the function $f(z) = z$ on the domain $\{x+iy \mid x \in [-1,1] , \; y \in [-1,1]\}$.

```python
f = lambda z : z
xbounds = (-1,1)
ybounds = (-1,1)
plt.title('f(z)=z')
plot_complex(f,xbounds,ybounds)
```

Plot the function $f(z) = \sqrt{z^2 + 1}$ on the domain $\{x+iy \mid x \in [-3,3] , \; y \in [-3,3]\}$.

```python
g = lambda z : np.sqrt(z**2+1)
xbounds = (-3,3)
ybounds = (-3,3)
plt.title('f(z)=sqrt(z**2+1) [angle]')
plot_complex(g,xbounds,ybounds)
```

## Problem 2

```python
def plot_complex_magnitude(f, xbounds, ybounds, res=401):
    """Plot the magnitude of the complex function f.
    INPUTS:
        f - A function handle. Should represent a function from C to C.
        xbounds - A tuple (xmin, xmax) describing the bounds on the real part of the domain.
        ybounds - A tuple (ymin, ymax) describing the bounds on the imaginary part of the domain.
        res - A scalar that determines the resolution of the plot (number of points
            per side). Defaults to 401.
    
    OUTPUTS:
        graph of the magnitude of the function f(z)
    """
    #raise NotImplementedError("Problem 2 Incomplete")
    x = np.linspace(xbounds[0],xbounds[1],res)
    y = np.linspace(ybounds[0],ybounds[1],res)
    X,Y = np.meshgrid(x,y)
    Z = f(X + 1j*Y)
    plt.pcolormesh(X,Y,np.absolute(Z),cmap='Blues')
    plt.axis([xbounds[0],xbounds[1],ybounds[0],ybounds[1]])
    plt.show()
```

Plot $f(z) = \sqrt{z^2 + 1}$ as in Problem 1, but plot the magnitude instead of the angle

```python
g = lambda z : np.sqrt(z**2+1)
xbounds = (-3,3)
ybounds = (-3,3)
plt.title('f(z)=sqrt(z**2+1) [magnitude]')
plot_complex_magnitude(g,xbounds,ybounds)
```

Plot the complex functions $z^2$, $z^3$, and $z^4$:

```python
f = lambda z : z**2
g = lambda z : z**3
h = lambda z : z**4
plt.subplot(311)
plt.title('f(z)=z^2')
plot_complex(f,(-1,1),(-1,1))
plt.subplot(312)
plt.title('f(z)=z^3')
plot_complex(g,(-1,1),(-1,1))
plt.subplot(313)
plt.title('f(z)=z^4')
plot_complex(h,(-1,1),(-1,1))
```

```python
f = lambda z : z**3-1j*z**4-3*z**6
plot_complex(f,(-1,1),(-1,1))
```

Compare it to your plot of $z^3$, especially near the origin. Based on these plots, what can you learn about the zeros of a function from its graph?

The zeros are where a fair number of different colors meet.  So this function has 4 visible zeros in our domain.  Also, there function cycles around each zero as many times as the greatest power of that zero in the function.

Plot the following complex functions:

Plot $f(z) = \sqrt{z}$

Now plot $f(z) = -\sqrt{z}$ to see the "other square root" of $z$. 

```python
f = lambda z : -np.sqrt(z)
plt.title('f(z)=-sqrt(z)')
plot_complex(f,(-1,1),(-1,1))
```

Describe why these two plots look the way they do.

Since the multiplicity is $\frac{1}{2}$, it only gets half way through the angles for one full rotation.  Depending on which way you start will change the angle that is represented (by the color) to the other end of the spectrum, which results in problems whenever $0$ is included.
