```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import scale
from sklearn.svm import SVC
```

### Problem 1
Load and look at the data

```python
cancer = load_breast_cancer()
data = pd.DataFrame(data=cancer.data,columns=[cancer.feature_names])
data.describe()
```

### Problem 2
Normalize the data so that all features have mean 0 and variance 1.

```python
x = scale(cancer.data)
y = cancer.target
norm_data = pd.DataFrame(data=x,columns=[cancer.feature_names])
norm_data.describe()
```

### Problem 3
Apply SVC with an RBF kernel.  Try all values of C and gamma in the set {2^k | k =-5..+10} (so 121 total pairs (C,gamma)) and for each of these, do 3-fold cross validation.  Identify which pair gives the best cross-validation score.

```python
cs = [2**k for k in range(-5,6)]
C = []
G = []
S = []
for c in cs :
    for gamma in cs :
        clf = SVC(C=c,gamma=gamma)
        scores = cross_val_score(clf, x, y, cv=3)
        average = np.mean(scores)
        C.append(c)
        G.append(gamma)
        S.append(average)
i = S.index(max(S))
print("Best score was C={}, gamma={} with an avgerage score of {}".format(C[i],G[i],S[i]))
```

### Problem 4
Repeat #3 with a polynomial kernel.  Now you also need to search over possible choices of degree (say, 1..5) and coeff0 (say 0, 1, -1), for a total of 121 x 5 x 3 different models.

```python
cs = [2**k for k in range(-5,11)]
C = []
G = []
K = []
C0 = []
S = []
for c in cs :
    for gamma in cs :
        for k in range(1,6) :
            for coeff0 in [0,-1,1] :
                # The comp had extreme difficulty computing these
                if (gamma,k,coeff0) == (16,4,-1) and c >= 32 :
                    pass
                elif (gamma,k,coeff0) == (8,4,-1) and c >= 512 :
                    pass
                else :
                    clf = SVC(C=c,kernel='poly',degree=k,coef0=coeff0,gamma=gamma)
                    scores = cross_val_score(clf, x, y, cv=3)
                    average = np.mean(scores)
                    C.append(c)
                    G.append(gamma)
                    K.append(k)
                    C0.append(coeff0)
                    S.append(average)
```

```python
i = S.index(max(S))
print('Best combo is C={}, gamma={}, degree={}, coef0={}, with an average score of {}.'.format(C[i],G[i],K[i],C0[i],S[i]))
```
