#### Tree Data

```python
tree = data("trees")
road = data("road")
b_d_rates = data("birthdeathrates")
heart = data("heart")
b_weight = data("birthwt")
```

```python
tree.plot()
plt.title("All data")
plt.show()
```

```python
gir = tree["Girth"]
x = np.linspace(min(gir),max(gir),len(gir))
plt.hist(gir,bins=x)
plt.ylabel("# of trees")
plt.xlabel("Girth (in)")
plt.show()
```

Distribution of tee girth.  4 of the trees had a girth of 11.

```python
hei = tree["Height"]
x = np.linspace(min(hei),max(hei),len(hei))
plt.hist(hei,bins=x)
plt.ylabel("# of trees")
plt.xlabel("Height (ft)")
plt.title("Tree Height")
plt.show()
```

This shows the distribution of tree height.  Apparently, 5 of the trees have a height of around 80 ft.

```python
tree.plot(kind="scatter", x="Girth", y="Height")
plt.title("Girth vs. Height")
plt.show()
```

This graph plots the girth of a tree vs. it's height, showing possible correlations between the two.  In this data set, they're not as related as I would have thought, but it does seem that shorter trees definitely have a smaller girth.

#### Road Data

```python
road.plot()
plt.title("All data")
plt.show()
```

```python
road.loc[road.index, ["deaths","drivers"]].plot(kind="barh",title="Deaths vs # of Drivers")
plt.tight_layout()
plt.show()
```

Plots the number of deaths against the number of drivers for each state.  In most states, the more drivers, the greater number of deaths, which probably is just a correlation of having a higher population.

```python
road.loc[road.index, ["temp","fuel"]].plot(kind="barh",title="Avg Temp vs Fuel Consumption")
plt.tight_layout()
plt.show()
```

Average temperature in January against fuel consumption.  While the units don't make since together, the thought was to see if temperature had a significant impact on fuel consumption, which doesn't seem to be the case.

#### Birth/Death Rate Data

```python
b_d_rates.plot()
plt.title("All data")
plt.show()
```

```python
birth = b_d_rates["birth"]
death = b_d_rates["death"]
b_d_rates["growth"] = birth-death
b_d_rates.drop(["birth","death"],axis=1).plot()
plt.title("Population Growth")
plt.show()
```

Plots the growth of each country.  It seems that there are 3 countries with a much larger increase in population size than the others, and only one where the growth was negative (indicating a smaller population size).

#### Heart Data

```python
heart.plot()
plt.title("All data")
plt.show()
```

```python
heart["duration"] = heart["stop"] - heart["start"]
heart.drop(["start","stop","event","age","year","surgery","transplant","id"],axis=1).plot(title="Duration of Episodes")
plt.show()
```

Duration of "episodes" which, best I can tell, is when a transplant patient is having a negative reaction to the surgery.  It seems that at least half the time, if not more, there is some kind of an episode.

```python
heart.plot(kind='scatter',x="year",y="transplant",title='Accepted to Recieved Transplant')
unserviced = len(heart['year']) - np.count_nonzero(heart['transplant'])
print("Number of accepted patients without a transplant when data collected : {}".format(unserviced))
plt.show()
```

While this graph doesn't seem very usueful, it helps us get an idea of how frequently they accept someone to the transfer list, and how frequently they're able to perform a transplant surgery.  Also, through the calculation above, it seems that there are roughly 103 patients still waiting for a transplant at the time of the collection of this data.

#### Birthweight Data

```python
b_weight.plot()
plt.title("All data")
plt.show()
```

```python
plt.hist(b_weight['bwt'])
plt.xlabel('Grams')
plt.ylabel('Number of Children')
plt.title('Birth Weight Distribution')
plt.show()
```

Distribution of birthweight seems to be more or less normally distributed.

```python
b_weight.loc[b_weight['smoke'] == 0].drop(['low','age','lwt','race','smoke','ptl','ht','ui','ftv'],axis=1).plot(kind='box',title="Non-smoking mothers")
b_weight.loc[b_weight['smoke'] == 1].drop(['low','age','lwt','race','smoke','ptl','ht','ui','ftv'],axis=1).plot(kind='box',title="Smoking mothers")
plt.show()
```

It would appear that while a non-smoking mother can have just as lightweight a child as a smoking mother, smoking mothers have a lower average of birth weight for their newborns.  Thus, smoking increases the likelihoood that your child will be born with a lower body weight.
